**2.** Using the [Penn World Table](https://www.rug.nl/ggdc/docs/pwt90.xlsx), conduct growth accounting of Hong Kong, Singapore, South Korea and Taiwan. Examine data from 1975 to 2014. Use rgdpna and rkna as output and capital stock measures. Create a table (similar to Table 7.3 of Weil (2013)). Briefly comment on the table.

```python
# Download the data
download_data = True
if download_data:
    data = pd.read_excel('https://www.rug.nl/ggdc/docs/pwt90.xlsx', sheet_name='Data')
    # Set country codes and years as index
    data = data.set_index(['countrycode','year'])

# These are the countries on which we will do growth accounting
country_codes = ['HKG','SGP','KOR','TWN']
```

```python
key = 'emp'
# income and capital per worker for this question
data['income per worker (3)'] = data['cgdpe']/data[key] 
data['capital per worker (3)'] = data['ck']/data[key]

# income and capital per worker for question 4
data['income per worker (4)'] = data['rgdpna']/data[key]
data['capital per worker (4)'] = data['rkna']/data[key]

alpha = 1/3
data['factors of production (4)'] = data['capital per worker (4)']**alpha*data['hc']**(1-alpha)
```

** Solution:** First, calculate growth rates per year from 1975 to 2014,
$$ g_y = \left(\frac{y_{2014}}{y_{1975}} \right)^{1/39}-1 $$
$$ g_k = \left(\frac{k_{2014}}{k_{1975}} \right)^{1/39}-1 $$
$$ g_h = \left(\frac{h_{2014}}{h_{1975}} \right)^{1/39}-1 $$

Assuming $y=A k^\alpha h^{1-\alpha}$, then $g_y = g_A + \alpha g_k + (1-\alpha)g_h$. Therefore, growth rate of productivity is $$g_A = g_y - \alpha g_k - (1-\alpha)g_h$$.

Lastly, contribution of productivity growth to income growth is equal to $g_A/g_y$, contribution of physical capital growth to income growth is $\alpha g_k/g_y$, and contribution of human capital growth to income growth is $(1-\alpha)g_h/g_y$

The variables we use are

rgdpna: Real GDP at constant 2011 national prices (in mil. 2011US\$)

rkna: Capital stock at constant 2011 national prices (in mil. 2011US\$)


```python
# create growth rate of factors as described above

data_q4=data.loc[(country_codes,(1975,2014)),:][['hc','income per worker (4)','capital per worker (4)','factors of production (4)']]
```

```python
# here is our data, as specified in the question we use rgdpna and rkna as measures of output and capital
# stock per worker since these are more suitable to calculate growth rates within a country
data_q4.round(decimals=2)
```

```python
# now calucalate the factor growth rates and productivity growth rates using formulas described above
def growth_rates(group):
    return (group.iloc[-1]/group.iloc[-0])**(1/39)-1

data_growth=data_q4.groupby('countrycode').apply(growth_rates)

alpha=1/3
data_growth['productivity'] = (data_growth['income per worker (4)'] - alpha*data_growth['capital per worker (4)']
                               - (1-alpha)*data_growth['hc'])
```

```python
data_growth.columns= 'human capital','income per worker','capital per worker','factors of production', 'productivity'
data_growth.round(decimals=3)
```

```python
data_growth.plot(kind='bar')
plt.legend(loc='best',frameon=False)
```

```python
# now let's calculate contribution of each factor to income growth
weight = np.array([(1/(1-alpha),1/alpha,1,1)])
weight.shape = (1,4)
income = data_growth['income per worker'].values
income.shape = (4,1)
data_growth_cont = data_growth[['human capital','capital per worker','factors of production','productivity']]/income.dot(weight)
data_growth_cont.round(2)
```

```python
data_growth_cont[['factors of production','productivity']].plot(kind='bar')
plt.legend(loc='best',frameon=False)
plt.ylim(ymax=.9)
plt.xlabel('Country')
plt.title('Contributions to growth')
```

Paul Krugman:
>  Communist growth rates were certainly impressive, but not magical. The rapid growth in output could be fully explained by rapid growth in inputs: expansion of employment, increases in education levels, and, above all, massive investment in physical capital. Once those inputs were taken into account, the growth in output was unsurprising--or, to put it differently, the big surprise about Soviet growth was that when closely examined it posed no mystery.

> Rapid Soviet economic growth was based entirely on one attribute: the willingness to save, to sacrifice current consumption for the sake of future production.

>  Economic growth that is based on expansion of inputs, rather than on growth in output per unit of input, is inevitably subject to diminishing returns.

>[Growth Accounting]: IT IS A TAUTOLOGY that economic expansion represents the sum of two sources of growth. On one side are increases in "inputs": growth in employment, in the education level of workers, and in the stock of physical capital (machines, buildings, roads, and so on). On the other side are increases in the output per unit of input; such increases may result from better management or better economic policy, but in the long run are primarily due to increases in knowledge.

> Sustained growth in a nation's per capita income can only occur if there is a rise in output per unit of input.

> In a famous estimate, MIT Professor Robert Solow concluded that technological progress has accounted for 80 percent of the long-term rise in U.S. per capita income, with increased investment in capital explaining only the remaining 20 percent.

> But what they actually found was that Soviet growth was based on rapid growth inputs--end of story. The rate of efficiency growth was not only unspectacular, it was well below the rates achieved in Western economies. Indeed, by some estimates, it was virtually nonexistent.

> The newly industrializing countries of Asia, like the Soviet Union of the 1950s, have achieved rapid growth in large part through an astonishing mobilization of resources. Once one accounts for the role of rapidly growing inputs in these countries' growth, one finds little left to explain, Asian growth, like that of the Soviet Union in its high-growth era, seems to be driven by extraordinary growth in inputs like labor and capital rather than by gains in efficiency.

> Consider, in particular, the case of Singapore. Between 1966 and 1990, the Singaporean economy grew a remarkable 8.5 percent per annum, three times as fast as the United States; per capita income grew at a 6.6 percent rate, roughly doubling every decade. This achievement seems to be a kind of economic miracle. But the miracle turns out to have been based on perspiration rather than inspiration: Singapore grew through a mobilization of resources that would have done Stalin proud. The employed share of the population surged from 27 to 51 percent. The educational standards of that work force were dramatically upgraded: while in 1966 more than half the workers had no formal education at all, by 1990 two-thirds had completed secondary education. Above all, the country had made an awesome investment in physical capital: investment as a share of output rose from 11 to more than 40 percent.

> There are two answers to that question. First, while many authors have written of an "Asian system"--a common denominator that underlies all of the Asian success stories--the statistical evidence tells a different story. Japan's growth in the 1950s and 1960s does not resemble Singapore's growth in the 1970s and 1980s. Japan, unlike the East Asian "tigers," seems to have grown both through high rates of input growth and through high rates of efficiency growth. Today's fast growth economics are nowhere near converging on U.S. efficiency levels, but Japan is staging an unmistakable technological catch-up.

>  On the contrary, Kim and Lau find "no apparent convergence between the technologies" of the newly industrialized nations and the established industrial powers; Young finds that the rates in the growth of efficiency in the East Asian "tigers" are no higher than those in many advanced nations.
