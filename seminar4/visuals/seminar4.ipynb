```python
# First import the packages we'll need
# Numpy is for numerical analysis
import numpy as np
# Pandas is for data storage and manipulation
import pandas as pd
# Matplotlib and seaborn are for plotting
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Sets the theme for our plots
import qeds
qeds.themes.mpl_style();
colors=plt.rcParams["axes.prop_cycle"].by_key()["color"]
colors.remove('#fec630')
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('./GERD_SOF.csv')
```

```python
df.head()
```

```python
select = ((df.Sector == 'Total intramural') &
         (df.Measure == '2010 Dollars - Constant prices and PPPs')&
         (df.COUNTRY== 'GBR'))
```

```python
df2 = df[select].copy()
```

```python
sectors = df2['Source of funds'].unique()
colors_ = dict(zip(sectors,colors))
```

```python
fig, ax = plt.subplots()
for sector in sectors:
    df3 = df2[df2['Source of funds'] == sector].copy()
    ax.plot(df3.year,df3.Value/1000,color=colors_[sector],linewidth=2)
    if sector == 'Higher education sector':
        loc_ = (df3.iloc[-1].year+1,df3.iloc[-1].Value/1000-1)
    else:
        loc_ = (df3.iloc[-1].year+1,df3.iloc[-1].Value/1000)        
    ax.annotate(sector,loc_,color=colors_[sector])
ax.spines['left'].set_visible(False)  
ax.spines['right'].set_visible(False)  
ax.spines['top'].set_visible(False)  
ax.spines['bottom'].set_visible(False)
ax.set_xlabel('Year')
ax.set_ylabel('Billion USD, 2010-prices')
plt.savefig('./rd_expenditures_uk.svg',bbox_inches='tight')
```

```python
GDP = pd.read_csv('./GDP.csv')
```

```python
select = ((GDP.TRANSACT == 'B1_GA') &
         (GDP.MEASURE == 'C'))
```

```python
df_gdp = GDP[select].copy()
```

```python
df_gdp.rename(columns={'Value':'GDP',
                      'LOCATION':"COUNTRY"},inplace=True)
```

```python
df.head()
```

```python
select = ((df.Measure == 'National Currency')&
         (df['Sector']=='Total intramural'))
```

```python
df_rd = df[select][['COUNTRY','Country','Sector','Measure','Source of funds','year','Value']].copy()
```

```python
df_rd.rename(columns={'year':'Year'},inplace=True)
```

```python
df_rd = df_rd.merge(df_gdp[['COUNTRY','Country','Year','GDP']],
            on=['COUNTRY','Country','Year'],how='left')
```

```python
df_rd['Value_per_GDP'] = df_rd['Value']/df_rd['GDP']
```

```python
sectors = df_rd['Source of funds'].unique()
colors_ = dict(zip(sectors,colors))
fig, ax = plt.subplots()
for sector in sectors:
    df3 = df_rd[(df_rd['Source of funds'] == sector)&(df_rd.COUNTRY=='GBR')].copy()
    ax.plot(df3.Year,df3.Value_per_GDP*100,color=colors_[sector],linewidth=2)
    if sector == 'Higher education sector':
        loc_ = (df3.iloc[-1].Year+1,df3.iloc[-1].Value_per_GDP*100-.06)        
    else:
        loc_ = (df3.iloc[-1].Year+1,df3.iloc[-1].Value_per_GDP*100)        
    ax.annotate(sector,loc_,color=colors_[sector])
ax.spines['left'].set_visible(False)  
ax.spines['right'].set_visible(False)  
ax.spines['top'].set_visible(False)  
ax.spines['bottom'].set_visible(False)
ax.set_xlabel('Year')
ax.set_ylabel('Percent')
plt.savefig('./rd_expenditures_gdp_uk.svg',bbox_inches='tight')
```

```python
select = ((df_rd['Source of funds']=='Total funding')&
         (df_rd.Sector=='Total intramural'))
```

```python
countries = ['GBR','FRA','DEU','CHN','JPN']
colors_ = dict(zip(countries,colors))
```
