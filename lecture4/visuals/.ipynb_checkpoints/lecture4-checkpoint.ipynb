```python
df = pd.read_excel('https://www.rug.nl/ggdc/docs/pwt91.xlsx',sheet_name='Data')
```

```python
df.year.unique()
```

```python
df_fundamentals = (df[df.year.isin(np.arange(1970,2018))]
                   .groupby(['countrycode','country'])['csh_i'].mean().reset_index().rename(columns={'csh_i':'Investment rate'}))
```

```python
df.sort_values(['countrycode','year'],inplace=True)
```

```python
df_emp_growth = (df[df.year.isin([1970,2017])]
                 .groupby(['countrycode']).apply(lambda x: (x['emp'].values[1]/x['emp'].values[0])**(1/47)-1)
                 .reset_index().rename(columns={0:'Employment growth'}))
```

```python
df_pop_growth = (df[df.year.isin([1970,2017])]
                 .groupby(['countrycode']).apply(lambda x: (x['pop'].values[1]/x['pop'].values[0])**(1/47)-1)
                 .reset_index().rename(columns={0:'Population growth'}))
```

```python
df_fundamentals= df_fundamentals.merge(df_emp_growth,on=['countrycode'],how='left').merge(df_pop_growth,on=['countrycode'],how='left')
```

```python
df['gdpPerworker'] = df['rgdpna']/df['emp']/1.0e6*1000
df['capPerworker'] = df['rnna']/df['emp']/1.0e6*1000
df['gdpPercap'] = df['rgdpna']/df['pop']/1.0e6*1000
df['capPercap'] = df['rnna']/df['pop']/1.0e6*1000
```

```python
df_fundamentals = df_fundamentals.merge(df[df.year == 2017][
    ['emp','countrycode','gdpPerworker','gdpPercap','pop','hc']],on='countrycode',how='left')
```

```python
df_fundamentals['Investment rate, USA'] = usa['Investment rate'].values[0]
df_fundamentals['Population growth, USA'] = usa['Population growth'].values[0]
df_fundamentals['Employment growth, USA'] = usa['Employment growth'].values[0]
df_fundamentals['hc, USA'] = usa['hc'].values[0]
df_fundamentals['gdpPerworker, USA'] = usa['gdpPerworker'].values[0]
df_fundamentals['gdpPercap, USA'] = usa['gdpPercap'].values[0]
```

```python
df_fundamentals.head()
```

$$y^\ast = A^{1/(1-\alpha)}\left(\frac{\gamma}{\delta+n} \right)^{\alpha/(1-\alpha)}h$$

```python
alpha=1/3
delta = .05
alpham = alpha/(1-alpha)
```

```python
df_fundamentals['rel_GDP_cap_pop']=((1/
                                          (delta+df_fundamentals['Population growth']))
                                         /(1/
                                          (delta+df_fundamentals['Population growth, USA'])))**alpham
```

```python
df_fundamentals['rel_GDP_cap_emp']=((1/
                                          (delta+df_fundamentals['Employment growth']))
                                         /(1/
                                          (delta+df_fundamentals['Employment growth, USA'])))**alpham
```

```python
df_fundamentals['rel_GDP_cap'] = df_fundamentals['gdpPercap']/df_fundamentals['gdpPercap, USA']
```

```python
df_fundamentals['rel_GDP'] = df_fundamentals['gdpPerworker']/df_fundamentals['gdpPerworker, USA']
```

```python
# var_names = ['rel_GDP_pred_inv','rel_GDP_pred_inv_emp','rel_GDP_pred_inv_emp_hc']
# var_name = var_names[0]
fig,ax = plt.subplots(figsize=(8,8))
ax.scatter(df_fundamentals['rel_GDP_cap_pop'],df_fundamentals['rel_GDP_cap'],
          sizes=df_fundamentals['pop'].values,color='k',alpha=.7)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.set_xlabel('Predicted income per capita relative to the US')
ax.set_ylabel('Actual income per capita relative to the US')
ticks = [.4,.8,1.2];
ax.set_xlim(0,1.6);
ax.set_ylim(0,1.6);
ax.set_xticks(ticks);
ax.set_yticks(ticks);


# ax.legend(loc='upper left',frameon=False)
plt.savefig('./rel_GDP_cap_pred_pop.svg',bbox_inches='tight')
```

```python
alpha = 1/3
delta = .05
n = 0.01
gamma = .2
f = lambda k: k**alpha
fdot = lambda k: gamma*k**alpha-(delta+n)*k
klist = [6.2]
```

```python
for t in range(100):
    k_prime = klist[t]+fdot(klist[t])
    klist.append(k_prime)
```

```python
klist[:3]
```

```python
ylist = np.array(klist)**alpha
```

```python
ylist[:3]
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,4))
ax[0].plot(np.arange(len(klist)),klist,'k',linewidth=2)
ax[1].plot(np.arange(len(klist)),ylist,'k',linewidth=2)
for i in range(2):
    ax[i].set_xlabel('Time')
    ax[i].spines['left'].set_visible(False)
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['top'].set_visible(False)
    ax[i].spines['bottom'].set_visible(False)
ax[0].set_title('Capital per worker, $k$')
ax[1].set_title('Output per worker, $y$')
#plt.savefig('./capital-output-simul.svg',bbox_inches='tight')
```
