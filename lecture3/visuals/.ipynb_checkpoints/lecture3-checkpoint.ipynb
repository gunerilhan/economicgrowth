In the Solow Model, $\dot{k} = \gamma f(k) - (\delta+n)k$.

Approximately, $k(t+1) \approx k(t) + \dot{k(t)}$, for $t=0,1,2,...$

```python
# Simulate the Solow Model

# Set parameters
A = 1 # productivity
h = 1 # human capital
alpha = 1/3 # capital elasticity of output
delta = .05 # depreciation rate
n = 0.01 # population growth rate
gamma = .2 # investment rate
f = lambda k: k**alpha # a way of defining simple functions, production function in this case
fdot = lambda k: gamma*k**alpha-(delta+n)*k # kdot : change in the capital per worker
klist = [1] # the list of k over time, with k(0)=1
T = 150 # number of time periods to simulate
```

```python
# Simulate T periods
for t in range(T):
    # next periods capital per worker = current k + kdot
    k_prime = klist[t]+fdot(klist[t]) 
    # append k_prime to our list of ks
    klist.append(k_prime)
```

Remember $ y=Ak^\alpha h^{1-\alpha}.$

```python
# After we calculated sequence of k over time, we can easily calculate 
# sequence of output per worker over time with the above formula
ylist = A*np.array(klist)**alpha*h**(1-alpha)
```

```python
# Plot graphs
fig,ax = plt.subplots(1,2,figsize=(12,4))
ax[0].plot(np.arange(len(klist)),klist,'k',linewidth=2)
ax[1].plot(np.arange(len(klist)),ylist,'k',linewidth=2)
for i in range(2):
    ax[i].set_xlabel('Time')
ax[0].set_title('Capital per worker, $k$')
ax[1].set_title('Output per worker, $y$')
plt.savefig('../files/capital-output-simul.svg',bbox_inches='tight')
```

### Quantitative Predictions of the Solow Model

Remember,

$$ \frac{y_i^{\ast}}{y_j^{\ast}}= \left(\frac{A_i}{A_j}\right)^{1/(1-\alpha)}\left(\frac{\gamma_i}{\gamma_j}\right)^{\alpha/(1-\alpha)}\left(\frac{\delta_j+n_j}{\delta_i+n_i}\right)^{\alpha/(1-\alpha)}\left(\frac{h_i}{h_j}\right).$$

To calculate the above ratio for the countries in our data set, we will need to find values of $\gamma$, $n$, and $h$. We will assume $\delta = 0.05$ and $A_i$ is equal to $A_j$ in this analysis. 

We will use the mean of the variable 'csh_i' from 1970 to 2019 as a proxy for the investment rate, $\gamma$.

We will use the mean of the varaible 'hc' from 1970 to 2019 as a proxy for human capital, $h$.

We will use average employment growth from 1970 to 2019 as a proxy for population growth, $n$. Remember, in the model, employment is equal to population and hence population growth is equal to employment growth. In the data, they are not equal. Here, we will use employment growth as a measure of $n$. We will calculate employment growth from the variable 'emp'.

```python
# Here are the descriptions of the variables we use:
print('csh_i = ',pwt_legend['csh_i'])
print('hc = ', pwt_legend['hc'])
print('emp = ',pwt_legend['emp'])
```

```python
# Select the years from 1970 to 2019
# For each country, calculate the avearge invesmtent rate with variable 'csh_i'
# rename 'csh_i' variable as 'Investment rate'
init_year = 1970 # the first year in the dataset
last_year = 2019 # the last year in the dataset
df_fundamentals = (df[df.year.isin(np.arange(init_year,last_year))]
                   .groupby(['countrycode','country'])['csh_i'].mean().reset_index().rename(
                       columns={'csh_i':'Investment rate'}))
```

```python
# Here is how df_fundamentals looks like
df_fundamentals.head()
```

```python
# Now calculate the employment growth
# First sort the original dataset
df.sort_values(['countrycode','year'],inplace=True)
# Calculate the employment growth rate
# average growth rate = (emp(2019)/emp(1970))^(1/(49))-1

# first select years 1970 amd 2019
# then apply the above formula for each country
# rename the new variable to Employment growth
df_emp_growth = (df[df.year.isin([init_year,last_year])]
                 .groupby(['countrycode']).apply(lambda x: (x['emp'].values[1]/x['emp'].values[0])
                                                 **(1/(last_year-init_year))-1)
                 .reset_index().rename(columns={0:'Employment growth'}))
```

```python
# merge investment rate data with employment growth data
df_fundamentals= df_fundamentals.merge(df_emp_growth,on=['countrycode'],how='left')
```

```python
# Calculate actual gdp per worker in thousand dollars
df['gdpPerworker'] = df['cgdpe']/df['emp']/1.0e6*1000
#df['capPerworker'] = df['rnna']/df['emp']/1.0e6*1000
```

```python
pwt_legend['cgdpe']
```

```python
# Add gdp per worker and human capital to df_fundamentals
df_fundamentals = df_fundamentals.merge(df[df.year == 2017][['emp','countrycode','gdpPerworker','hc']],on='countrycode',how='left')
```

```python
# since we will calculate every thing relative to the US
# select the usa data from df_fundamentals
usa = df_fundamentals.query("countrycode=='USA'")
```

```python
# Add USA data as seperate columns
df_fundamentals['Investment rate, USA'] = usa['Investment rate'].values[0]
df_fundamentals['Employment growth, USA'] = usa['Employment growth'].values[0]
df_fundamentals['hc, USA'] = usa['hc'].values[0]
df_fundamentals['gdpPerworker, USA'] = usa['gdpPerworker'].values[0]
```

```python
# here is how the data looks like
df_fundamentals.head()
```

$$y^\ast = A^{1/(1-\alpha)}\left(\frac{\gamma}{\delta+n} \right)^{\alpha/(1-\alpha)}h$$

```python
# define alpha
alpha=1/3
delta = .05
alpham = alpha/(1-alpha)
```

When $A_i=A_j$, $\delta_i=\delta_j$, $n_i=n_j$, and $h_i=h_j$
$$ \frac{y_i^{\ast}}{y_j^{\ast}}= \left(\frac{\gamma_i}{\gamma_j}\right)^{\alpha/(1-\alpha)}.$$

```python
# Calculate Solow predicted relative incomes with the above formula
df_fundamentals['rel_GDP_pred_inv']=(df_fundamentals['Investment rate']/df_fundamentals['Investment rate, USA'])**alpham
```

```python
# Calculate actual relative gdp per worker
df_fundamentals['rel_GDP'] = df_fundamentals['gdpPerworker']/df_fundamentals['gdpPerworker, USA']
```

When $A_i=A_j$, $\delta_i=\delta_j$, and $h_i=h_j$

$$ \frac{y_i^{\ast}}{y_j^{\ast}}= \left(\frac{\gamma_i}{\gamma_j}\right)^{\alpha/(1-\alpha)}\left(\frac{\delta_j+n_j}{\delta_i+n_i}\right)^{\alpha/(1-\alpha)}.$$

```python
# Calculate Solow predicted relative incomes with the above formula
df_fundamentals['rel_GDP_pred_inv_emp']=((df_fundamentals['Investment rate']/
                                          (delta+df_fundamentals['Employment growth']))
                                         /(df_fundamentals['Investment rate, USA']/
                                          (delta+df_fundamentals['Employment growth, USA'])))**alpham
```

When $A_i=A_j$ and $\delta_i=\delta_j$

$$ \frac{y_i^{\ast}}{y_j^{\ast}}= \left(\frac{\gamma_i}{\gamma_j}\right)^{\alpha/(1-\alpha)}\left(\frac{\delta_j+n_j}{\delta_i+n_i}\right)^{\alpha/(1-\alpha)}\left(\frac{h_i}{h_j}\right).$$

```python
# Calculate Solow predicted relative incomes with the above formula
df_fundamentals['rel_GDP_pred_inv_emp_hc'] = (df_fundamentals['rel_GDP_pred_inv_emp']*
                                              df_fundamentals['hc']/df_fundamentals['hc, USA'])
```

```python
# Now calculate correlations of various predictions with the actual value
df_fundamentals[['rel_GDP_pred_inv','rel_GDP_pred_inv_emp','rel_GDP_pred_inv_emp_hc','rel_GDP']].corr().round(decimals=2)
```

```python
# Drop observations with missing values
df_fundamentals.dropna(subset=['rel_GDP_pred_inv','rel_GDP_pred_inv_emp','rel_GDP_pred_inv_emp_hc',
                               'emp','rel_GDP'],inplace=True)
```

```python
var_names = ['rel_GDP_pred_inv','rel_GDP_pred_inv_emp','rel_GDP_pred_inv_emp_hc']
labels = ['$\gamma$','$\gamma,n$','$\gamma,n,h$']
for i in range(3):
    fig,ax = plt.subplots(figsize=(6,6))
    for j in range(i+1):
        ax.scatter(df_fundamentals[var_names[j]],df_fundamentals['rel_GDP'],
                  sizes=df_fundamentals['emp'].values,color=colors[2][j-1],alpha=.7,label=labels[j])
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')
    ticks = [0,.4,.8,1.2]
    ax.set_xlim(-0.2,1.3)
    ax.set_ylim(-0.2,1.3)
    ax.set_xticks(ticks)
    ax.set_yticks(ticks)
    ax.legend(loc='upper left',frameon=False)
    plt.savefig(f'../files/{var_names[i]}.svg',bbox_inches='tight')
```
